#### Import SUL

```python
# importando dados rais
df_rais_vinculos_sul = pd.read_csv('2019/RAIS_VINC_PUB_SUL.txt', sep=';' , encoding='latin-1', decimal=',', usecols = cols_name)

## Cria coluna ano
df_rais_vinculos_sul['ano'] = 2019

# # Reorganizado posição das colunas. 
df_rais_vinculos_sul = df_rais_vinculos_sul[[
            'ano',
            'Município',
            'Nacionalidade',
            'Raça Cor',
            'Sexo Trabalhador',
            'Idade',
            'Faixa Etária',
            'Escolaridade após 2005',
            'Vínculo Ativo 31/12',
            'CBO Ocupação 2002',
            'CNAE 2.0 Classe', 
            'CNAE 2.0 Subclasse',
            'Natureza Jurídica',
            'Tamanho Estabelecimento', 
            'Tipo Estab',  
            'Tipo Admissão',
            'Tipo Vínculo',
            'Vl Remun Média Nom',
            'Vl Remun Média (SM)' ]]

# Filtrar trabalhadores ativos em 31/12
df_rais_vinculos_sul = df_rais_vinculos_sul[(df_rais_vinculos_sul['Vínculo Ativo 31/12'] == 1)]

## Gera planilha de dados
df_rais_vinculos_sul.to_csv('2019/rais_2019/df_rais_vinculos_sul.csv', index=False)

## limpeza de memória
del df_rais_vinculos_sul
gc.collect()
```
